```python
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_csv('AlphaMissense_hg38.tsv', sep='\t', skiprows=3)
```

```python
def pathogenicity_score(data, gene, variant):
    reduced_data = data[data['uniprot_id']==gene][data['protein_variant']==variant]
    score = np.mean(reduced_data['am_pathogenicity'])
    return score
```

```python
tree = {}

for iter, row in data_short.iterrows():
    uniprot = row['uniprot_id']
    variant = row['protein_variant']
    pathogenicity = row['am_pathogenicity']
    pclass = row['am_class']

    if uniprot not in tree:
        tree[uniprot] = {}
    if variant not in tree[uniprot]:
        tree[uniprot][variant] = []
    
    tree[uniprot][variant].append({'Pathogenicity_Score': pathogenicity, 'Class': pclass})

if iter % 10 == 0:
    print(iter)
```

```python
# Convert the tree to a JSON string
json_tree = json.dumps(tree, indent=4)
print(json_tree)
```
